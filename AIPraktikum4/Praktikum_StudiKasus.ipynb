Membuat Data Studi Kasus

```python
import pandas as pd

data = pd.DataFrame({
  'IPK': [3.5, 2.1, 3.8, 2.5, 3.2, 2.0, 3.6, 2.7],
  'kehadiran': [90, 60, 95, 70, 85, 50, 92, 75],
  'jam_belajar': [15, 5, 20, 8, 12, 4, 18, 9],
  'jumlah_organisasi': [1, 0, 2, 1, 1, 0, 2, 1],
  'lulus': [1, 0, 1, 0, 1, 0, 1, 0]
})
print(data)
```

Memisahkan Fitur dan Target

```python
x = data[['IPK', 'kehadiran', 'jam_belajar', 'jumlah_organisasi']]
y = data['lulus']
```

Data training dan testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

Melatih model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=500)
model.fit(x_train_scaled, y_train)
print("Model berhasil dilatih.")
```

Prediksi

```python
prediksi = model.predict(x_test_scaled)
print("hasil prediksi:", prediksi)
```

Menghitung Akurasi

```python
from sklearn.metrics import accuracy_score
akurasi = accuracy_score(y_test, prediksi)
print("Akurasi model:", akurasi)
```

Uji Data baru

```python
mahasiswa_baru = pd.DataFrame({
    'IPK': [3.4],
    'kehadiran': [70],
    'jam_belajar': [10],
    'jumlah_organisasi': [1]
})
mahasiswa_baru_scaled = scaler.transform(mahasiswa_baru)
hasil = model.predict(mahasiswa_baru_scaled)
if hasil[0] == 1:
    print("Prediksi: Mahasiswa kemungkinan LULUS")
else:
    print("Prediksi: Mahasiswa kemungkinan TIDAK LULUS")
```
