IMPORT LIBRARY AI

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

LOAD DATASET

```python
digits = datasets.load_digits()
print("Jumlah data: ", len(digits.data))
print("Jumlah fitur: ", len(digits.data[0]))
```

TAMPILKAN GAMBAR

```python
plt.imshow(digits.images[0], cmap='gray')
plt.title("contoh gambar angka")
plt.show()
```

LATIHAN MODEL SEDERHANA

```python
from sklearn.linear_model import LogisticRegression
X = digits.data
y = digits.target
model = LogisticRegression(max_iter=10000)
model.fit(X, y)
print("Modul berhasil dilatih")
```

MODEL AKAN MENEBAK

```python
prediksi = model.predict(X[0:10])
print("Prediksi: ", prediksi)
print("jawaban asli: ", y[0:10])
```
